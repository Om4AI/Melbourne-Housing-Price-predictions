## ***Get and extract data into a Pandas DataFrame***

```python
import sklearn
import zipfile
from zipfile import ZipFile
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with ZipFile("melbourne_housing.zip",'r') as zip:
    zip.extractall()
```

```python
melbourne_data = pd.read_csv("melb_data.csv")
print(type(melbourne_data))
melbourne_data.head()
```

```python
# Basic statistical values for the Melbourne Dataset

melbourne_data.describe()
```

```python
# All features in the dataset
melbourne_data.columns
```

```python
melbourne_data.dropna(axis=0)
```

## ***Select target and features from dataset***

```python
# Select the target for the predictions

y = melbourne_data.Price
type(y)
```

```python
# Select features

melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
x = melbourne_data[melbourne_features]
x.describe()
```

```python
x.head()
```

## ***Decision Tree model***

```python
from sklearn.tree import DecisionTreeRegressor

decision_model = DecisionTreeRegressor(random_state=42)
decision_model.fit(x, y)
```

```python
# Making initial predictions
print("Making predictions for the following data: ")
print(x.head())
print(melbourne_data[['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude','Price']].head())
print("\n The precitions: ")
print(decision_model.predict(x.head()))
```

## ***Model Evaluation: Mean Absolute Error (MAE): In-sample scores***

```python
from sklearn.metrics import mean_absolute_error

predictions = decision_model.predict(x)
mean_absolute_error(y, predictions)
```

### ***Spliting the data into train & test sets***

```python
from sklearn.model_selection import train_test_split

train_x, val_x, train_y, val_y = train_test_split(x, y , random_state=42)

# Prepare and fit the model on the new datasplit
decision_model = DecisionTreeRegressor(random_state=42)
decision_model.fit(train_x, train_y)

# Make predictions on the validation set
val_predictions = decision_model.predict(val_x)
print(mean_absolute_error(val_y, val_predictions))
```

#### ***The above is the cost error for the validation set (This is way too much)***

### ***Function for comparing the depth of the tree to the MAE***

```python
def get_mae(max_leaf_nodes, train_x, val_x, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=42)
    model.fit(train_x, train_y)
    preds_val = model.predict(val_x)
    return (mean_absolute_error(val_y, preds_val))
```

### ***Find the best size for max leaf nodes for the Decision Tree***

```python
# Test the function for different values of tree depths
min_mae = 1000000000

for max_leaf_nodes in [5, 25, 50, 100, 250, 500]:
    mae = get_mae(max_leaf_nodes, train_x, val_x, train_y, val_y)
    if (mae < min_mae):
        min_mae = mae
        best_nodes_size = max_leaf_nodes
    print("Max Leaf nodes: ",max_leaf_nodes,"\t\tMean Absolute error: ",mae)
print("\n\n Best choice for the maximum number of nodes is: ",best_nodes_size)
```

#### ***From the options listed we can say that 500 is the optimal choice***

### ***Final model for Decision Tree***

```python
# Define the model
decision_final_model = DecisionTreeRegressor(max_leaf_nodes=best_nodes_size)

# Fit on all the data
decision_final_model.fit(x, y)

# Evaluate the model
final_preds = decision_final_model.predict(val_x)
print("Final MAE: ",mean_absolute_error(val_y, final_preds))
```

#### ***Earlier we had the MAE much higher, now that we found the better max leaf nodes so the MAE has decreased***

## ***Random Forest model***

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest_model = RandomForestRegressor(random_state=42)
forest_model.fit(train_x, train_y)
forest_preds = forest_model.predict(val_x)
print("Mean absolute error for Random Forests model: ",mean_absolute_error(val_y, forest_preds))
```

#### ***The above result is good for the initial default parameters; we can also further tune the parameters to improve the performance***
